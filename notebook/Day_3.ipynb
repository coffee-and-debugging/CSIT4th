```python
# Create a new file named "hello.ipynb"
# 'x' mode is used to create a file and will raise an error if the file already exists
file = open("data.txt", "x")

# Close the file after creation to free system resources
file.close()
```

```python
# Open the file "data.txt" in read mode
file = open("data.txt", "r")

# Read the entire content of the file
content = file.read()

# Print the file content
print(content)

# Close the file to free system resources
file.close()
```

```python
# Open (or create) the file "data.txt" in write mode
# 'w' mode will overwrite existing content if the file already exists
file = open("data.txt", "w")

# Write text into the file
file.write("Heheh")

# Close the file to save changes and free resources
file.close()
```

```python
# Open the file "data.txt" in append mode
# 'a' mode adds content at the end without deleting existing data
file = open("data.txt", "a")

# Add a new line to the file
file.write("\nNew line added")

# Close the file to save changes
file.close()
```

```python
# Try block: code that may cause an error
try:
    # Take input from user and convert it to integer
    num = int(input("Enter a number: "))
    
    # Perform division
    print(10 / num)

# Except block: runs if any error occurs
except:
    print("Error occurred!")

# Finally block: always runs no matter what
finally:
    print("This block will always be executed!")
```

```python
# Define a class named Student
class Student:
    # Class attribute (shared by all objects of the class)
    name = "Ram"

# Create an object of the Student class
s1 = Student()

# Access and print the attribute using the object
print(s1.name)
```

```python
# Define a class named Student
class Student:
    # Constructor method to initialize object attributes
    def __init__(self, name):
        # Assign value to instance variable
        self.name = name

# Create an object of Student class and pass value "Ram"
s1 = Student("Ram")

# Print the name attribute of the object
print(s1.name)
```

```python
# Define a class named Student
class Student():
    # Method inside the class
    def show(self):
        print("Hello World!")

# Create an object of Student class
s1 = Student()

# Call the method using the object
s1.show()
```

```python
# Base class (Parent class)
class Animal:
    # Method in parent class
    def sound(self):
        print("Animal can make sounds")

# Derived class (Child class) inheriting Animal
class Dog(Animal):
    pass  # No additional methods or properties

# Create object of Dog class
d = Dog()

# Call inherited method from Animal class
d.sound()
```

```python
# Simple Calculator Program

# Print title
print("Simple Calculator")

# Take input from user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print("\n")

# Display menu options
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

# Take user's choice
choice = int(input("Enter your choice: "))

# Perform operation based on choice
if choice == 1:
    result = num1 + num2
    print("Addition:", result)

elif choice == 2:
    result = num1 - num2
    print("Subtraction:", result)

elif choice == 3:
    result = num1 * num2
    print("Multiplication:", result)

else:
    result = num1 / num2
    print("Division:", result)
```
