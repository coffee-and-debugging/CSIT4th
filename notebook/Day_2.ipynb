## If-Else Statement

### If
The `if` keyword starts the conditional logic by testing a condition and executes a block of code if the condition is true.

### Elif
The `elif` keyword means "if the previous condition is not true, then check this condition".

### Else
The `else` keyword runs when none of the above conditions are true.

```python
# Assign values to variables
a = 50
b = 100

# Check conditions
if a == b:
    # Runs if a and b are equal
    print("a and b are equal")

elif a > b:
    # Runs if a is greater than b
    print("a is greater than b")

else:
    # Runs if none of the above conditions are true
    print("b is greater than a")
```

## Python Loops

Python has two primitive loop commands:

- while loops  
- for loops  

---

## The while Loop

With the `while` loop, we can execute a set of statements as long as a condition is true.

---

## The break Statement

The `break` statement is used to stop the loop even if the condition is still true.

---

## The continue Statement

The `continue` statement is used to skip the current iteration and move to the next iteration.

```python
i=1
while i<6:
    print(i)
    i+=1
```

## Python For Loops

A `for` loop is used for iterating over a sequence (such as a list, tuple, dictionary, set, or string).

```python
# A list containing different types of items
fruits = ['apple', 'banana', 'cherry', 1]

# Loop through each element in the list
for x in fruits:
    # Print each element one by one
    print(x)
```

```python
# Loop through each character in the string "banana"
for x in "banana":
    # Print each character one by one
    print(x)
```

## Insert Items in a List

To insert a new item in a list **without replacing existing values**, we use the `insert()` method.

```python
# Create a list of fruits
fruit = ['apple', 'banana', 'cherry']

# Insert "orange" at index 1 (second position)
fruit.insert(1, "orange")

# Print the updated list
print(fruit)
```

## Append Items

To add an item to the **end of a list**, we use the `append()` method.

```python
# Create a list of fruits
fruit = ['apple', 'banana', 'cherry']

# Add "orange" at the end of the list
fruit.append("orange")

# Print the updated list
print(fruit)
```

## Extend List

To add elements from another list into the current list, we use the `extend()` method.

```python
# Create first list of fruits
list1 = ["apple", "banana", "cherry"]

# Create second list of fruits
list2 = ["mango", "pineapple", "papaya"]

# Extend list1 by adding all elements of list2
list1.extend(list2)

# Print the updated list1
print(list1)
```

## Remove Specified Item

The `remove()` method removes the specified item from a list.

```python
# Create a list of fruits
fruit = ["apple", "banana", "cherry"]

# Remove "banana" from the list
fruit.remove("banana")

# Print the updated list
print(fruit)
```

## Remove Specified Index

The `pop()` method removes an item at a specified index.

```python
# Create a list of fruits
fruit = ["apple", "banana", "cherry"]

# Remove the last item from the list using pop()
fruit.pop()

# Print the updated list
print(fruit)
```

## Clear the List

The `clear()` method removes all items from a list.

The list still exists, but it becomes empty.

```python
# Create a list of fruits
fruit = ["apple", "banana", "cherry"]

# Remove all items from the list
fruit.clear()

# Print the empty list
print(fruit)
```

## Loop Through a List

You can loop through the items of a list using a `for` loop and `while` loop

```python
# Create a list of fruits
fruits = ["apple", "banana", "cherry"]

# Loop through each item in the list
for x in fruits:
    # Print each item
    print(x)
```

```python
# Create a list of fruits
fruits = ["apple", "banana", "cherry"]

# Loop through the list using index values
for i in range(len(fruits)):
    # Print each element using its index
    print(fruits[i])
```

```python
# Create a list of fruits
fruits = ["apple", "banana", "cherry"]

# Initialize counter variable
i = 0

# Loop while i is less than the length of the list
while i < len(fruits):
    # Print the current element
    print(fruits[i])

    # Increment i by 1
    i = i + 1
```
