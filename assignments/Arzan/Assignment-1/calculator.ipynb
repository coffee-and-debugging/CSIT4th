```python
#CALCULATOR PROJECT


while True:
    num1=int(input("Enter first number:  "))
    num2=int(input("Enter second number: "))
    print(f"Here are following choices:")
    print("1.addition")
    print("2.subtraction")
    print("3.multiplication")
    print("4.division")
    print("5.modulo")

    choose=int(input("enter your choice: "))
    if choose==1:
        print(f"addition:{num1+num2}")
    elif choose==2:
        print(f"subtraction:{num1-num2}")
    elif choose==3:
        print(f"multiplication:{num1*num2}")
    elif choose==4:
        if num2!=0:
            print(f"division:{num1/num2}")
        
        else:
            print(f"division:undefined")
    
    elif choose==5:
        if num2!=0:
            print(f"modulo:{num1%num2}")
        
        else:
            print(f"modulo:undefined")
    else:
        print(f"invalid choice")
    again=input(f"Do you want to use calculator again:(yes/no)")

    if again=="no":
        print("calculator is closed")
        break
```
